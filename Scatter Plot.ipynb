```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1=pd.read_csv('data_2 - Growth Rates.csv')
```

```python
df2=pd.read_csv('1980.csv')
```

```python
df3=pd.merge(df1,df2,on='Country')
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,7))
sns.regplot(y='Average Rate of growth', x='log(Real GDP per capita)', data=df3, ci=None)
plt.xlabel('log(Real GDP per capita) for 1980')
plt.title('156 Countries',fontsize=17)
```
