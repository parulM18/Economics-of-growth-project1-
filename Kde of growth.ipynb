```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math
import scipy.stats as stats
```

```python
df=pd.read_csv('data_2 - Growth Rates.csv')
```

```python
sns.set_style("darkgrid")
P=sns.FacetGrid(data=df,height=6,aspect=1.5, legend_out=True, palette="Set1").map(sns.kdeplot, "Average Rate of growth")
plt.ylabel('Density',fontsize=15)
plt.xlabel('Average Rate of growth of log real gdp per capita',fontsize=15)
plt.legend()
```

```python
sns.set_style("darkgrid")
P=sns.FacetGrid(data=df,height=6,aspect=1.5, legend_out=True).map(sns.kdeplot, "Average Rate of growth")
plt.ylabel('Density',fontsize=15)
plt.xlabel('Average Rate of growth of log real gdp per capita',fontsize=15)
mean=df['Average Rate of growth'].mean()
median=df['Average Rate of growth'].median()
mode= -0.001786
plt.axvline(mean, color='r', linestyle='-.', label='Mean')
plt.axvline(median, color='g', linestyle='-.', label='Median')
plt.axvline(mode, color='y', linestyle='-.', label='Mode')
plt.legend() 
plt.show()
```
