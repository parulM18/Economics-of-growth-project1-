```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("data_3 - growth (1).csv")
```

```python
df.head()
```

```python
df = df[1:]
```

```python
df.head()
```

```python
plt.grid()
P=sns.FacetGrid(data=df,height=6,aspect=1.5, legend_out=True, palette="Set1").map(sns.kdeplot, "Solow Residual")
plt.ylabel('Density',fontsize=15)
plt.xlabel('Solow Residual',fontsize=15)
plt.legend()
plt.style.use("dark_background")
```

```python
plt.grid()
P=sns.FacetGrid(data=df,height=6,aspect=1.5, legend_out=True).map(sns.kdeplot, "Solow Residual")
plt.ylabel('Density',fontsize=15)
plt.xlabel('Solow Residual',fontsize=15)
#plt.title('Kernel density plot of Solow Residual',fontsize=23, fontweight="bold")
mean=df['Solow Residual'].mean()
median=df['Solow Residual'].median()
mode= 0.01111111111
plt.axvline(mean, color='r', linestyle='--', label='Mean')
plt.axvline(median, color='g', linestyle='-.', label='Median')
plt.axvline(mode, color='y', linestyle='-.', label='Mode')
plt.legend(fontsize= 'x-large') 
plt.grid(True, linewidth=0.5, color='#27272d', linestyle='-')
plt.show()
#plt.style.use(color = "#27272d")
#plt.grid(True, linewidth=0.5, color='#27272d', linestyle='-')
```
