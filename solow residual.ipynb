```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("data_3 - growth (1).csv")
```

```python
df = df[1:]
```

```python
#plt.grid()
P=sns.FacetGrid(data=df,height=6,aspect=2, legend_out=True).map(sns.kdeplot, "Solow Residual", color='#c0504d')
sns.set(rc={'axes.facecolor':'#27272d', 'figure.facecolor':'#27272d'})

COLOR = 'white'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR

plt.ylabel('Density',fontsize=15, color='w')
plt.xlabel('Solow Residual',fontsize=15, color='w')
#plt.title('Kernel density plot of Solow Residual',fontsize=23, fontweight="bold")
mean=df['Solow Residual'].mean()
median=df['Solow Residual'].median()
mode= 0.01111111111
plt.axvline(mean, color='#f79646', linestyle='--', label='Mean')
plt.axvline(median, color='#4f81bd', linestyle='--', label='Median')
plt.axvline(mode, color='#9bbb59', linestyle='--', label='Mode')
plt.legend(fontsize= 'x-large') 
plt.grid(True, linewidth=0.5, color='w', linestyle='-')
plt.show()
#plt.grid()
#plt.style.use(color = "#27272d")
#plt.grid(True, linewidth=0.5, color='#27272d', linestyle='-')
```
